```python
import numpy as np

##calculo nomra infinito  norma_inf = np.linalg.norm(A, ord = np.inf)

def metodo_jacobi(A, b, x0, n_max = 15, eps = 10**(-5)):
    
    it = 1
    n = len(A)
    x = np.zeros(n)
    error = 1
    while (it <= n_max and error > eps):
        for i in range(n):
            suma = 0
            for j in range(n):
                if i != j:
                    suma += A[i,j]*x0[j]
                    
            x[i] = (b[i] - suma)/A[i,i]
            
        print(x)
        error = np.linalg.norm(x-x0, ord = np.inf)
        x0[:] = x
        it += 1

    if it > n_max :
        print("Numero maximo de iteraciones alcanzado")
        return x

    print(f"Convergencia alcanzada en la iteracion {it-1}")
    return x    
            
            
A = np.array([[6,-2,1],[-2,7,2],[1,2,-5]])
B = np.array([4,1,-4.5])
n = len(A)
x0 = np.zeros(n)
x = metodo_jacobi(A,B,x0,10, 10**(-2))
print(x)

    
    
```

```python
import numpy as np
def metodo_sor(A, b, x0, w, n_max = 10,eps = 10**(-3)):
    n = len(A)
    x_ant = x0.copy()
    it = 1
    error = 1
    
    while( it <= n_max and error > eps):
        for i in range(n):
            suma = 0
            for j in range(n):
                if i != j:
                    suma += A[i,j]*x0[j]

            x0[i] = w*((b[i] - suma)/A[i,i]) + (1-w)*x_ant[i]

        error = np.linalg.norm(x0 - x_ant, ord = np.inf)
        x_ant[:] = x0
        it += 1

    if(it > n_max):
        print("Numero maximo de iteraciones alcanzadas")
    else:
        print(f"Convergencia alcanzada en la iteracion {it-1}")

    return x0
        
            
    
A = np.array([[6,-2,1.0],[-2,7,2.0],[1,2,-5.0]])
B = np.array([4,1,-4.5])
n = len(A)
x0 = np.zeros(n)
w = 1.2
x = metodo_sor(A,B,x0,w,15,10**(-2))
print(x)  

















```
