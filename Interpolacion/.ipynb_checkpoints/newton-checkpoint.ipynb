Forma de Newton 
Idea : dado el polinomio de grado n-1 : Pn-1, poder calcular Pn reutilizando lo anterior.
Pn(x) = Pn-1(x) + C(x)
C(x) = Pn(x) - Pn-1(x), C(x) debe ser de orden n. 

Newton propuso el siguiente polinomio
Pn(x) = a0 + a1(x-x0) + a2(x-x0)(x-x1) + a3(x-x0)(x-x1)(x-x2) + ... + an-1(x-x0)(x-1)...(x-xn-2)+ an(x-x0)(x-x1)...(x-xn-1)

Plantenado Pn-1(x)
Pn-1(x) = a0 + a1(x-x0) + a2(x-x0)(x-x1) + a3(x-x0)(x-x1)(x-x2) + ... + an-1(x-x0)(x-1)...(x-xn-2)

Calculando C(x) tenemos:
C(x) = Pn(x) - Pn-1(x) = an(x-x0)(x-x1)...(x-xn-1)

C(x) tiene los n ceros  x0, x1, x2,...,xn-1 y por ser interpolante Pn(xn) = f(xn)
Despejando an obtenemos: 
an = Pn(xn) - Pn-1(xn) / (x-x0)(x-x1)...(x-xn-1)
an = f(xn) - Pn-1(xn) /  (x-x0)(x-x1)...(x-xn-1) 
an = F[x0,x1,x2,...,xn] n-esima diferencia divida de f en x0,x1,x2,...,x-2

Contrusccion del polinomio: 
Con un punto x0, f(x0) -> p0(x) = a0 = f(x0) -> F[x0] = f(x0)
Con 2 puntos: x0,f(x0)  x1,f(x1) -> p1(x1) = a0 + a1(x1-x0) = f(x1) 
despejando a1 = f(x1) - a0 / x1 - x0 -> a1 = f(x1) - f(x0) / x1 - x0 = f[x0,x1]

P1(x) = f(x0) + f[x0,x1](x-x0)

Pn(x) = Pn-1(x) + f[x0,...,xn](x-x0)...(x-xn-1)








```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def diferencias_divididas(x,y):
    d = y.copy()
    n = len(x)
    
    for i in range(1,n):
        for j in range(n-1, i-1, -1):
            d[j] = (d[j] - d[j-1]) / (x[j] - x[j-i])

    return d


def polinomio_newton(x,d):
    
    X = sp.Symbol('x')
    n = len(x)
    suma = d[0]
    
    for i in range(1,n): #1,2,3
        prod = 1
        for j in range(i): #2 -- 0,1
            prod *= (X - x[j])
            
        suma += d[i]*prod

    return sp.expand(suma)
    

## a0 + a1(x-x0) + a2(x-x0)(x-x1)

x = np.array([-3,-2,-1,0,1,2,3.0])
y = np.array([9,4,1,0,1,4,9.0])


x2 = np.array([-3,-2,-1,0,1,2.0])
y2 = np.array([9,4,1,0,1,4.0])

dif = diferencias_divididas(x2,y2)

print(dif)
pol = polinomio_newton(x,dif)
print(pol)

pol_f = sp.lambdify('x',pol)
print(f"{pol_f(1.5)}")

def graficar_polinomio_interpolante(x,y,f):
    a = x.min()
    b = x.max()
    
    xs = np.linspace(a,b,50)
    ys = f(xs)
    
    plt.scatter(x,y)
    plt.plot(xs,ys)
    plt.grid()

graficar_polinomio_interpolante(x,y,pol_f)

```

```python
###Esta implementacion es mejor que la anterior porque la ant tiene costo O(n^2)
###Mejor seria aplicar Horner

import numpy as np
import sympy as sp

def dif_div2(x,y):
    d = y.copy()
    n = len(x)
    for i in range(1,n+1):
        for j in range(n-1, i-1, -1):
            d[j] = (d[j] - d[j-1])/(x[j] - x[j-i])
            
    return d


def regla_termino_siguiente(x,y,x0):
    d = dif_div2(x,y)
    n = len(x)
    print(d)
    prod = d[n-1]
    X = sp.Symbol('x')
    for i in range(n):
        prod *= (X - x[i])

    termino_siguiente = sp.lambdify(X,prod)
    return termino_siguiente(x0)
    


def pol_newton2(x,y,x0):
    d = dif_div2(x,y)
    print(d)
    n = len(x)
    
    X = sp.Symbol('x')
    suma = d[0]
    prod = 1
    
    for i in range(1,n):
        prod *= (X - x[i-1])
        suma += d[i]*prod

    print(f"p(x)={sp.expand(suma)}")
    pol_f = sp.lambdify(X,suma)
    return pol_f(x0)


x = np.array([4,-4,7,6,2.0])
y = np.array([278,-242,1430,980,40.0])

xs = np.array([4,-4,7,6.0])
ys = np.array([278,-242,1430,980.0])

pol = pol_newton2(xs,ys,5)
print(f"p(5)={pol}")

print()

error = regla_termino_siguiente(x,y,5) 
print(f"error cometido={error}")
###El error cometido por  no usar ese punto.+

```
